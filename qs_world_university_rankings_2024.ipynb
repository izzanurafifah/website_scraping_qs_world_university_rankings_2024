```python
from bs4 import BeautifulSoup
import requests
```

```python
base_url = "https://www.topuniversities.com/rankings/endpoint?nid=3897789&page={}&items_per_page=100&tab="

total_pages = 15
all_universities = []

for page_number in range(total_pages):
  url = base_url.format(page_number)
  page = requests.get(url)
  json_data = page.json()
  json_data = json_data['score_nodes']
  all_universities.extend(json_data)
```

```python
print(all_universities)
```

```python
def get_title(titles):
  title_list = []
  for title in titles:
    title = title['title']
    title_list.append(title)
  return title_list

def get_region(regions):
  region_list = []
  for region in regions:
    region = region['region']
    region_list.append(region)
  return region_list

def get_country(countries):
  country_list = []
  for country in countries:
    country = country['country']
    country_list.append(country)
  return country_list

def get_city(cities):
  city_list = []
  for city in cities:
    city = city['city']
    city_list.append(city)
  return city_list

def get_overallscore(overallscores):
  overallscore_list = []
  for overallscore in overallscores:
    overallscore = overallscore['overall_score']
    overallscore_list.append(overallscore)
  return overallscore_list

def get_rank(ranks):
  rank_list = []
  for rank in ranks:
    rank = rank['rank']
    rank_list.append(rank)
  return rank_list
```

```python
titles = get_title(all_universities)
regions = get_region(all_universities)
countries = get_country(all_universities)
cities = get_city(all_universities)
overall_scores = get_overallscore(all_universities)
ranks = get_rank(all_universities)
```

```python
table = list(zip(ranks, titles, overall_scores, cities, countries, regions))
```

```python
import pandas as pd

header = ['Rank', 'University', 'Overall Score', 'City', 'Country', 'Region']

df = pd.DataFrame(table, columns=header)

df
```
